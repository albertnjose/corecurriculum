# Exercise 1: Create a new column that contains only the zipcode.

```python
data = []
with open ('yelp_academic_dataset_business.json') as f:
    for line in f:
        data.append(simplejson.loads(line))

# convert to dataframe
df = pd.DataFrame(data[:5000]) # only use first 5000 rows
df.head()
```

# Exercise 2: The table contains a column called 'categories' and each entry in this column is populated by a list. We are interested in those businesses that are restaurants. Create a new column 'restaurant_type' that contains a description of the restaurant based on the other elements of 'categories. That is, if we have '[Sushi Bars, Japanese, Restaurants]' in categories the 'restaurant_type will be '{'SushiBars': 1, 'Japanese': 1, 'Mexican': 0, ...}'

```python
categories = {}

# separate all rows with restaurants and keep category, assign 0 initially
# use iterrows for pandas DF
for i, row in df.iterrows():
    if 'Restaurants' in row['categories']:
        for element in row['categories']:
            if element != 'Restaurants':
                categories[element] = 0
                
```

```python
categories
```

```python
def category(row):
    d = dict(categories)
    if 'Restaurants' in row['categories']:
        for el in row['categories']:
            if el in d.keys():
                d[el] = 1
            else:
                d[el] = 0
    else:
        d = 'Remove'
        
    return d
```

```python
df['restaurant_type'] = df.apply(lambda x: category(x) , 1)
```

# Exercise 3: Lets clean the 'attributes' column. The entries in this column are dictionaries. We need to do two things:

1) Turn all the True or False values in the dictionary to 1 and 0.

2) There are some entries within dictionaries that are dictionaries themselves, lets turn the whole entry into just one dictionary, for example if we have

'{'Accepts Credit Cards': True, 'Alcohol': 'none','Ambience': {'casual': False,'classy': False}}'

then turn it into

'{'Accepts Credit Cards':1, 'Alcohol': 0, 'Ambience_casual': 0, 'Ambience_classy': 0}'.

There might be other entries like {'Price Range': 1} where the values are numerical so we might want to change that into {'Price_Range_1': 1}.

```python
df.attributes[:10]
```

```python
def clean_attributes(row):
    d = row['attributes']
    d1 = {}
    
    # iterate over the keys
    for element in d.keys():
        
        # if value boolean change to 1 and 0
        if type(d[element]) == bool:
            d1[element] = int(d[element])
        
        # join strings in one key
        elif type(d[element]) == str:
            label = '{}_{}'.format(element, d[element])
            d1[label] = 1
        
        # join integers in one key
        elif type(d[element]) == int:
            label = '{}_{}'.format(element, d[element])
            d1[label] = 1
        
        # nested dictionaries, new key is combined keys and value is 1 or 0 
        elif type(d[element]) == dict:
            for elem2 in d[element].keys():
                label = '{}_{}'.format(element, elem2)
                d1[label] = int(d[element][elem2])
        
        # edge case check
        else:
            print(element, d[element])

    return d1
```

```python
df['attributes_cleaned'] = df.apply(lambda x: clean_attributes(x), 1)
```

# 5. Create a table with the average star review for a business from the review table.

```python
data_review = []

# load the review dataset
with open('yelp_academic_dataset_review.json') as f:
    for line in f:
        data_review.append(simplejson.loads(line))
        
df_review = pd.DataFrame(data_review[:5000]) # same approach as before
df_review.head()
```

```python
# groupby business id because it's unique and then average the stars and reset index
avg_rtg = df_review.groupby('business_id')['stars'].mean().reset_index()
```

```python
# set the columns
avg_rtg.columns = ['business_id', 'star_avg']
```

```python
avg_rtg.head()
```

```python
# merge from the original data set and new avg rating dataframe with left join
df_review = df_review.merge(avg_rtg, how='left', on='business_id')
df_review.head()
```
